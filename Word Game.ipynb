```python
import random
import sys
import warnings

warnings.filterwarnings('ignore')

print('-----------------------------')
print('-------Jumble Bubble---------')
print('-----------------------------')

while True:
    try:
        playing = input('Do you want to play a game? (yes/no) ')

        if playing.lower() == 'yes':
            print('Okay! Let\'s Play')
        elif playing.lower() == 'no':
            print('Thank you for playing')
            sys.exit()
        else:
            print('Invalid input. Please enter either "yes" or "no".')
            continue

        words = ['ironman', 'captain_america', 'thor', 'hulk', 'wanda', 'black_widow', 'hawk_eye', 'spiderman', 'vision',
                 'dr.strange', 'venom', 'captain_marvel', 'dormamu']

        word = random.choice(words)

        jumble = ''.join(random.sample(word, len(word)))
        print(jumble)

        chance = 5
        while chance != 0:
            print('The word is', jumble)

            guess = input('Enter your word: ').lower()

            if guess == word:
                print('Correct guess')
                print('You won!')
                print()
                break
            else:
                chance -= 1
                print('Incorrect guess')
                print('The remaining chances are', chance)
                print()
        else:
            print('All your chances are exhausted')
            print('You lost! ')
            print('The correct word is', word)

        play_again = input('Do you want to play again? (yes/no) ')
        if play_again.lower() == 'no':
            print('Thank you for playing')
            break  # Exit the loop and end the program

    except SystemExit:
        pass
```
